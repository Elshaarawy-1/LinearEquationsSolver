```python
import numpy as np
from mpmath import mp, matrix



class JacobiSolver:
    def __init__(self, A, b, itr=10, tolerance=0.00001, x=None, sf=5):
        self.A = A
        self.b = b
        self.itr = itr
        self.tolerance = tolerance
        self.x = mp.matrix(x)
        self.sf = sf

    def matrix_without_diagonal(self, A):
        Dia = mp.matrix(len(A))
        for i in range(len(A)):
            Dia[i, i] = A[i, i]
            A[i, i] = 0
        return A, Dia

    def check_zero_diagonal(self, matrix):
        return any(matrix[i, i] == 0 for i in range(len(matrix)))

    def solve(self):
        mp.dps = self.sf

        R, D = self.matrix_without_diagonal(self.A)
        R = mp.matrix(R)
        D = mp.matrix(D)

        for i in range(self.itr):
            print(f"Iteration {i+1}:")
            x_old = self.x.copy()
            print(f"x_old = {x_old}")

            for itr in range(len(self.x)):
                dot_product = mp.fdot(R[itr, :], x_old)
                self.x[itr] = (self.b[itr] - dot_product) / D[itr, itr]
                abs_rel_error = abs((self.x[itr] - x_old[itr]) / self.x[itr])
                print(f"x{itr} = {self.x[itr]} and Absolute relative error: {abs_rel_error}")

            err = np.abs((np.array(self.x.tolist()) - np.array(x_old.tolist())) / np.array(self.x.tolist()))
            if np.all(err < self.tolerance):
                print(f"\nConverged after {i + 1} iterations.")
                break

        return self.x

# Example usage:

A = mp.matrix([[-8, 1],
               [4, 2]])
B = mp.matrix([15.5, 5])

x = mp.matrix([[1],
               [1]])
x = JacobiSolver(A, B, 50, 0.1, x=x).solve()

print(x)
```
